# **3.Data Wrangling**

**Data Wrangling Code**

```python

# Fill missing values in 'Rating' with the median rating
df['Rating'].fillna(df['Rating'].median(), inplace=True)

# Drop rows where critical columns have missing values, e.g., 'App Name', 'App Id', etc.
df.dropna(subset=['App Name', 'App Id', 'Category', 'Installs', 'Released'], inplace=True)


```

```python
# Convert the 'Installs' column to string, handle NaN values, and then clean the data
df['Installs'] = df['Installs'].astype(str)
df['Installs'] = df['Installs'].str.replace(',', '', regex=False)
df['Installs'] = df['Installs'].str.replace('+', '', regex=False)
df['Installs'] = df['Installs'].replace('nan', '0')
df['Installs'] = df['Installs'].astype(int)
```

```python
# Convert the 'Price' column to string and then clean the data
df['Price'] = df['Price'].astype(str)
df['Price'] = df['Price'].str.replace('$', '', regex=False)
df['Price'] = df['Price'].astype(float)
```

```python
# Handle 'Size': Convert size to MB (if size in KB, convert to MB), and handle 'Varies with device' as missing
df['Size'] = df['Size'].replace('Varies with device', None)
df['Size'] = df['Size'].str.replace('M', '').str.replace('k', '')
df['Size'] = pd.to_numeric(df['Size'],errors='coerce')
df['Size'] = df['Size'].apply(lambda x: x / 1024 if 'k' in str(x) else x)
```

```python
# Convert 'Released' and 'Last Updated' columns to datetime
df['Released'] = pd.to_datetime(df['Released'], errors='coerce')
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
```

```python
# Convert 'Free' to boolean
df['Free'] = df['Free'].apply(lambda x: True if x == 'True' else False)
```

```python
# Replace missing 'Developer Website', 'Developer Email', and 'Privacy Policy' with 'No Information'
df['Developer Website'].fillna('No Information', inplace=True)
df['Developer Email'].fillna('No Information', inplace=True)
df['Privacy Policy'].fillna('No Information', inplace=True)
```

**What all manipulations have you done and insights you found?**

Answer:
*   Missing value in the 'Rating' column are filled with the median rating.
*   rows with missing values in these columns('App Name','App Id','Category','Installs','Released') are removed
*   Removes commas and plus signs from 'Installs', then convert it to an integer type
*   Removes dollar signs from 'Price' and convert the values to float type.
*   Replace 'Varies with device' with None,remove 'M' and 'K',and convert the size to numeric
*   converts 'Released' and 'Last Updated' columns to datetime format
*   converts the 'Free' column to a boolean type
*   Replaces missing values with 'No information' in developer-related fields








# **4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables**

**1.Univariate Analysis**

**Chart-1 : Distribution of Ratings**

```python
df['Rating'].plot(kind='hist',bins=20,color='skyblue')
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

**Chart-2 : App categories**

```python
df['Category'].value_counts().plot(kind='bar', color='orange')
plt.title('Number of Apps in Each Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

**Chart-3 : Box Plot of App sizes**

```python
df['Size'].plot(kind='box', color='green')
plt.title('Distribution of App Sizes')
plt.show()
```

**2.Bivariate Analysis**

**Chart-1 : Rating vs Rating Count**

```python
df.plot(kind='scatter', x='Rating', y='Rating Count', color='blue')
plt.title('Rating vs. Rating Count')
plt.show()
```

**Chart-2 : Average Rating by Category**

```python
df.groupby('Category')['Rating'].mean().plot(kind='bar', color='purple')
plt.title('Average Rating by Category')
plt.ylabel('Average Rating')
plt.show()
```

**Chart-3 : Rating vs Installs**

```python
df.plot(kind='scatter', x='Installs', y='Rating', color='blue', alpha=0.5)
plt.title('Rating vs Installs')
plt.xlabel('Number of Installs')
plt.ylabel('Rating')
plt.show()
```

**3.Multivariate Analysis**

**Chart-1 : Pair plot for Rating,Price and Installs**

```python
sns.pairplot(df[['Rating', 'Price', 'Installs']],diag_kind='kde')
plt.show()
```

**Chart-2 : Violin Plot for Ratings and Category**

```python
sns.violinplot(x='Category', y='Rating', hue='Free', split=True, data=df)
plt.title('Ratings by Category and Free/Paid Apps')
plt.xticks(rotation=90)
plt.show()
```

**Chart-3 : Rating vs Installs with Price as Bubble Size**

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Rating'], df['Installs'], s=df['Price'].fillna(0), alpha=0.5, c='blue', edgecolors='w', linewidth=0.5)
plt.title('Rating vs Installs with Price as Bubble Size')
plt.xlabel('Rating')
plt.ylabel('Installs')
plt.show()
```
